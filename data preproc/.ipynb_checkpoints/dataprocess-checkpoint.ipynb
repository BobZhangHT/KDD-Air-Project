```python
# import
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline 
```

```python
# set data directory
root_dir=os.getcwd()+'/'
dir_17_aq=root_dir+'beijing_17_18_aq.csv'
dir_17_me=root_dir+'beijing_17_18_meo.csv'

# load data by pandas
df_aq=pd.read_csv(dir_17_aq)
df_me=pd.read_csv(dir_17_me)
```

# Air Quality Data Preprocess

```python
print('The number of data is',len(df_aq['O3']))
df_aq
```

```python
df_aq.describe()
```

```python
# print time steps
time_aq,ind_aq=np.unique(df_aq['utc_time'],return_inverse=True)
print('There are',len(time_aq),'time steps in total')
print(time_aq)
print(ind_aq)
```

```python
# average over all stations
pollution=np.zeros(len(time_aq))
for i in range(len(time_aq)):
    pollution[i]=np.nanmean(df_aq[ind_aq==i]['PM2.5'])
# create new df
df1_aq=pd.DataFrame({
    'utc_time':time_aq,
    'PM2.5':pollution
})
df1_aq
```

# Meteorological Data Preprocess

```python
print('The number of data is',len(df_me['utc_time']))
df_me
```

```python
# set outliers of wind direction as nan
df_me['wind_direction']=df_me['wind_direction'].apply(lambda x: np.nan if x>360 else x)
df_me.describe()
```

```python
# print time steps
time_me,ind_me=np.unique(df_me['utc_time'],return_inverse=True)
print('There are',len(time_me),'time steps in total')
print(time_me)
print(ind_me)
```

```python
def weather_mode(seq):
    seq=list(seq)
    seq_class=np.unique(seq)
    freq=np.zeros(len(seq_class))
    for i in range(len(seq_class)):
        freq[i]=seq.count(seq_class[i])
    return(seq_class[np.argmax(freq)])
```

```python
# aggregate the results
temperature,pressure,humidity,wind_direction,wind_speed,weather=list(),list(),list(),list(),list(),list()
for i in range(len(time_me)):
    temperature.append(np.nanmedian(df_me[ind_me==i]['temperature']))
    pressure.append(np.nanmedian(df_me[ind_me==i]['pressure']))
    humidity.append(np.nanmedian(df_me[ind_me==i]['humidity']))
    wind_direction.append(np.nanmedian(df_me[ind_me==i]['wind_direction']))
    wind_speed.append(np.nanmedian(df_me[ind_me==i]['wind_speed']))
    weather.append(weather_mode(df_me[ind_me==i]['weather']))
```

```python
# create new df
df1_me=pd.DataFrame({
    'utc_time':time_me,
    'temperature':temperature,
    'pressure':pressure,
    'humidity':humidity,
    'wind_direction':wind_direction,
    'wind_speed':wind_speed,
    'weather':weather
})
df1_me
```

```python
# same the processed data
df1_me.to_csv(root_dir+'me_17.csv')
```

# Merge the data

```python
# data merge
left=df1_aq
right=df1_me
df_merge=pd.merge(left, right, how='right', on='utc_time')
df_merge=df_merge.sort_values(by='utc_time',ascending=True)
df_merge.index=np.arange(0,len(df1_me['utc_time']))
```

```python
df_merge
```

```python
df_merge.to_csv(root_dir+'merge_17_original.csv')
```

# NaN Imputation V1
I here make use of external [datasets](http://beijingair.sinaapp.com) to fill the NaN partially.

```python
# locate the NaNs
n=len(df_merge['utc_time'])
ind_nan=np.arange(0,n)
ind_nan=ind_nan[df_merge['PM2.5'].isna()]
print(ind_nan[:10])
np.sum(df_merge.isna())
```

```python
# set directories of external data folders
lib17_fdr='beijing_20170101-20171231/'
lib18_fdr='beijing_20180101-20180602/'
```

```python
def time_extract(utc_time):
    date=utc_time[:4]+utc_time[5:7]+utc_time[8:10]
    hour=utc_time[11:13]
    return date,hour
```

```python
nan_seq=list()
for ind in ind_nan:
    utc_time=df_merge['utc_time'][ind]
    date,hour=time_extract(utc_time)

    if utc_time[3]=='7':
        lib_dir=root_dir+lib17_fdr+'beijing_all_'+date+'.csv'
    else:
        lib_dir=root_dir+lib18_fdr+'beijing_all_'+date+'.csv'

    df_lib=pd.read_csv(lib_dir)
    # avoid error in lib
    if df_lib.shape[1]!=38:
        nan_seq.append(np.nan)
        continue
    
    df_lib=df_lib[df_lib['hour']==int(hour)]
    
    # avoid missing data in lib
    if df_lib.shape[0]==0:
        nan_seq.append(np.nan)
    else:
        nan_seq.append(np.nanmean(list(df_lib.iloc[0,3:])))
```

```python
df_merge['PM2.5'][ind_nan]=nan_seq
```

```python
# show the rest missing part of the data
np.sum(df_merge.isna())
```

```python
# save
df_merge.to_csv(root_dir+'merge_17_v1.csv')
```

# NaN Imputation V2
[The median of medians](https://biendata.com/forum/view_post_category/170) method will be introduced to resolve remained missing data.

## PM2.5

```python
ind_nan=np.arange(0,n)
ind_nan=ind_nan[df_merge['PM2.5'].isna()]
print(ind_nan[:10])
```

```python
# median of medians
def MedoMeds(df,colname,ind,win_seq=list([3,6,12,24,36,48,60,72,84,96,108,120,144,168,264,336])):
    # df : data frame
    # colname : column name of target variable
    # ind : index of variable
    med_seq=list()
    for i in range(len(win_seq)):
        win=win_seq[i]
        med_seq.append(np.nanmedian(df[colname][np.max((ind-win-1),0):(ind-1)]))
    return np.nanmedian(med_seq)
```

```python
nan_seq=list()
for ind in ind_nan:
    nan_seq.append(MedoMeds(df_merge,'PM2.5',ind))
df_merge['PM2.5'][ind_nan]=nan_seq
```

```python
np.sum(df_merge.isna())
```

## Wind speed

```python
ind_nan=np.arange(0,n)
ind_nan=ind_nan[df_merge['wind_speed'].isna()]
print(ind_nan[:10])

nan_seq=list()
for ind in ind_nan:
    nan_seq.append(MedoMeds(df_merge,'wind_speed',ind,win_seq=list([3,6,12,24,36,48])))
df_merge['wind_speed'][ind_nan]=nan_seq

np.sum(df_merge.isna())
```

## Wind Direction

```python
ind_nan=np.arange(0,n)
ind_nan=ind_nan[df_merge['wind_direction'].isna()]
print(ind_nan[:10])

nan_seq=list()
for ind in ind_nan:
    nan_seq.append(MedoMeds(df_merge,'wind_direction',ind,win_seq=list([3,6,12,24,36,48])))
df_merge['wind_direction'][ind_nan]=nan_seq

np.sum(df_merge.isna())
```
