```python
def uniq_aver(df_aq):
    # unique 
    time_aq,ind_aq=np.unique(df_aq['utc_time'],return_inverse=True)
    # average over all stations
    pollution=np.zeros(len(time_aq))
    for i in range(len(time_aq)):
        pollution[i]=np.nanmean(df_aq[ind_aq==i]['PM2.5'])
    # create new df
    df1_aq=pd.DataFrame({
        'utc_time':time_aq,
        'PM2.5':pollution
    })
    return df1_aq
```

```python
# unify the dataframe
df1_aq_p1=uniq_aver(df0_aq_p1)
df1_aq_p2=uniq_aver(df0_aq_p2)
```

## Meteological Data Concat

```python
# load data
df0_me_p1=pd.read_csv(root_dir+'beijing_201802_201803_me.csv')
df0_me_p2=pd.read_csv(root_dir+'bj_me_180331-180531.csv')

# change colname
colname4p2=list(str('i'))
colname4p2.extend(list(df0_me_p1.columns))
df0_me_p2.columns=colname4p2
```

```python
# show 
# df0_me_p1
df0_me_p2
```

```python
def weather_mode(seq):
    seq=list(seq)
    seq_class=np.unique(seq)
    freq=np.zeros(len(seq_class))
    for i in range(len(seq_class)):
        freq[i]=seq.count(seq_class[i])
    return(seq_class[np.argmax(freq)])

def uniq_aggr(df_me):
    # set outliers of variables as nan
    df_me['temperature']=df_me['temperature'].apply(lambda x: np.nan if x>=999999 else x)
    df_me['pressure']=df_me['pressure'].apply(lambda x: np.nan if x>=999999 else x)
    df_me['wind_speed']=df_me['wind_speed'].apply(lambda x: np.nan if x>=999999 else x)
    df_me['wind_direction']=df_me['wind_direction'].apply(lambda x: np.nan if x>=999999 else x)
    df_me['humidity']=df_me['humidity'].apply(lambda x: np.nan if x>=999999 else x)

    #  time steps
    time_me,ind_me=np.unique(df_me['utc_time'],return_inverse=True)

    # aggregate the results
    temperature,pressure,humidity,wind_direction,wind_speed,weather=list(),list(),list(),list(),list(),list()
    for i in range(len(time_me)):
        temperature.append(np.nanmedian(df_me[ind_me==i]['temperature']))
        pressure.append(np.nanmedian(df_me[ind_me==i]['pressure']))
        humidity.append(np.nanmedian(df_me[ind_me==i]['humidity']))
        wind_direction.append(np.nanmedian(df_me[ind_me==i]['wind_direction']))
        wind_speed.append(np.nanmedian(df_me[ind_me==i]['wind_speed']))
        weather.append(weather_mode(df_me[ind_me==i]['weather']))

    # create new df
    df1_me=pd.DataFrame({
        'utc_time':time_me,
        'temperature':temperature,
        'pressure':pressure,
        'humidity':humidity,
        'wind_direction':wind_direction,
        'wind_speed':wind_speed,
        'weather':weather
    })
    
    return df1_me
```

```python
df1_me_p1=uniq_aggr(df0_me_p1)
df1_me_p2=uniq_aggr(df0_me_p2)
```

# NaN Imputation V1

```python
def time_extract(utc_time):
    date=utc_time[:4]+utc_time[5:7]+utc_time[8:10]
    hour=utc_time[11:13]
    return date,hour

# locate the NaNs
n=len(df_merge['utc_time'])
ind_nan=np.arange(0,n)
ind_nan=ind_nan[df_merge['PM2.5'].isna()]

# set directories of external data folders
lib18_fdr='beijing_20180101-20180602/'

nan_seq=list()
for ind in ind_nan:
    utc_time=df_merge['utc_time'][ind]
    date,hour=time_extract(utc_time)

    if utc_time[3]=='7':
        lib_dir=root_dir+lib17_fdr+'beijing_all_'+date+'.csv'
    else:
        lib_dir=root_dir+lib18_fdr+'beijing_all_'+date+'.csv'

    df_lib=pd.read_csv(lib_dir)
    # avoid error in lib
    if df_lib.shape[1]!=38:
        nan_seq.append(np.nan)
        continue
    
    df_lib=df_lib[df_lib['hour']==int(hour)]
    
    # avoid missing data in lib
    if df_lib.shape[0]==0:
        nan_seq.append(np.nan)
    else:
        nan_seq.append(np.nanmean(list(df_lib.iloc[0,3:])))
        
df_merge['PM2.5'][ind_nan]=nan_seq
```

# NaN Imputation V2

```python
ind_nan=np.arange(0,n)
ind_nan=ind_nan[df_merge['PM2.5'].isna()]
print(ind_nan[:10])
```

```python
# median of medians
def MedoMeds(df,colname,ind,win_seq=list([3,6,12,24,36,48,60,72,84,96,108,120,144,168,264,336])):
    # df : data frame
    # colname : column name of target variable
    # ind : index of variable
    med_seq=list()
    for i in range(len(win_seq)):
        win=win_seq[i]
        med_seq.append(np.nanmedian(df[colname][np.max((ind-win-1),0):(ind-1)]))
    return np.nanmedian(med_seq)
```

```python
nan_seq=list()
for ind in ind_nan:
    nan_seq.append(MedoMeds(df_merge,'PM2.5',ind))
df_merge['PM2.5'][ind_nan]=nan_seq
```
